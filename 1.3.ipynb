#### Q-1 Write a Pythin program to demonstrate the use os type casting constructors (int(), float(), str(), bool()):
####  - Take input from the user as a string
####  - Convert the string into an integer, a float, and a boolean
####  - Print the converted values along with their types

```python
val = input("Enter the string: ")

int_val = int(val)
float_val = float(val)
str_val = str(val)
bool_val = bool(val)

print("Integer value:", int_val, "  Type:", type(int_val))
print("Float value:", float_val, "  Type:", type(float_val))
print("String value:", str_val, "  Type:", type(str_val))
print("Boolean value:", bool_val, "  Type:", type(bool_val))
```

#### Q-4 Write a Python program to:
####  - Declare a variable of each datatype(integer, float, string, boolean, list, tuple, dictionary)
####  - Print the value, type(using type()), and memory address using(id()) of each variable.

```python
int_val = 5
float_val = 8.4
str_val = "Python"
bool_val = True
l1 = ["Apple", "Banana", "Mango"]
t1 = ("Apple", "Banana", "Mango")
d1 = {"key1":"value1", "key2":"value2"}

print("Integer =", int_val, "\tType: ",type(int_val), "\tMemory address:", id(int_val))
print("Float =", float_val, "\tType: ",type(float_val), "\tMemory address:", id(float_val))
print("String =", str_val, "Type: ",type(str_val), "\tMemory address:", id(str_val))
print("Boolean =", bool_val, "\tType: ",type(bool_val), "\tMemory address:", id(bool_val))
print("List =", l1, "\tType: ",type(l1), "\tMemory address:", id(l1))
print("Tuple =",t1, "\tType: ",type(t1), "\tMemory address:", id(t1))
print("Dictionary =",d1, "\tType: ",type(d1), "\tMemory address:", id(d1))
```

#### Q-1 Write a program to demonstrate different formating option in print()
####     - use sep to seperate values with a custom character
####     - use end to customize what appears at the end of a print() statement

```python
# sep 
print("Apple", "Banana", "Orange", "Guava", "Mange", sep = "|")

#end
print("Hello", end = " World")
```

#### Q-3 Perform addition, subtraction, multiplication, division, floor division, modulus and exponentiation on two numbers input ny the user.

```python
val1 = int(input("Enter the first value:"))
val2 = int(input("Enter the second value:"))

#addition
print("The addition of",val1, "and", val2, "=", val1 + val2)

#Subtraction
print("The subtraction of",val1, "and", val2, "=", val1 - val2)

#Multiplication
print("The multiplication of",val1, "and", val2, "=", val1 * val2)

#Division
print("The division of",val1, "and", val2, "=", val1 / val2)

#Floor division
print("The floor division of",val1, "and", val2, "=", val1 // val2)

#modulus
print("The modulus of",val1, "and", val2, "=", val1 % val2)

#exponentiation
print("The exponentiation of",val1, "and", val2, "=", val1 ** val2)
```

#### Q-6 Implement a program to demostrate logical operators (and, or, not) by asking the user for boolean inputs(e.g., true/false values).

```python
val1 = input("True or False:")
val2 = input("True or False:")

print("AND operator:", val1 and val2)
print("OR operator:", val1 or val2)
print("NOT operator:", not val1)
```

#### Q-7 Write a program to deminstrate assignment operators(=, +=, -=, *=, /=) using a single variable

```python
a = 10
print("The value of a :",a)
a += 5
print("now the value of a :",a)
a -= 3
print("now the value of a :", a)
a *= 10
print("now the value of a :", a)
a /= 12
print("now the value of a :", a)
```
